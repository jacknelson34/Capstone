```python

# Define your category labels
categories = {0: "Nale", 1: "Qufu", 2: "Erci", 3: "AirBubble", 4: "QufuStem"}

# List to store the summary data
summary_data = []

# Iterate through each file in file_paths
for file in file_paths:
    try:
        # Load the data from the current source file
        df = pd.read_table(file, header=None, delim_whitespace=True)

        # Count occurrences of each category
        counts = {category: (df.iloc[:, 0] == label).sum() for label, category in categories.items()}

        # Create a row with the file name and counts
        row = [file.split('/')[-1]] + list(counts.values())

        # Add the row to the summary data
        summary_data.append(row)

    except Exception as e:
        print(f"Error processing {file}: {e}")

# Create a DataFrame from the summary data
summary_df = pd.DataFrame(summary_data, columns=["Source File"] + list(categories.values()))

# Save the summary DataFrame to a new CSV file
summary_df.to_csv("summary_data.csv", index=True)

print("Summary CSV created successfully!")
```
